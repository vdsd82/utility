```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


```

```python
def party(numbere):
    # URL of the webpage to scrape
    df = pd.DataFrame(columns=["First Party", "Second Party", "Difference","Constituency"])
    url = "https://results.eci.gov.in/ResultAcGenMay2023/ConstituencywiseS10" + str(numbere) + ".htm?ac=" + str(numbere)

    # Send a GET request to the URL
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table containing the candidate data
    table = soup.find("table", {"style": "margin: auto; width: 100%; font-family: Verdana; border: solid 1px black;font-weight:lighter"})

    # Extract the rows from the table
    rows = table.find_all("tr")
    
    constituency = table.find("td", colspan="9").text.strip()

    # Create an empty list to store the candidate data
    candidate_data = []

    # Iterate over the rows, skipping the header rows
    for row in rows[1:]:
        # Extract the cells from the row
        cells = row.find_all("td")

        # Extract the candidate name, party name, and total votes if there are enough cells
        if len(cells) >= 7:
            candidate_name = cells[1].text.strip()
            party_name = cells[2].text.strip()
            total_votes = int(cells[5].text.strip())

            # Add the candidate data to the list
            candidate_data.append((candidate_name, party_name, total_votes))

    # Sort the candidate data based on total votes in descending order
    sorted_data = sorted(candidate_data, key=lambda x: x[2], reverse=True)
    
    
    top_2_candidates = sorted_data[1:3]

    # Calculate the difference between the top 2 candidates' total votes
    difference = top_2_candidates[0][2] - top_2_candidates[1][2]
    

    # Get the top 3 candidates
    top_3_candidates = sorted_data[1:4]    
    # Get the party names of the first and second candidates
    first_party_name = sorted_data[1][1] if len(sorted_data) > 0 else ""
    second_party_name = sorted_data[2][1] if len(sorted_data) > 1 else ""
    

    # Get the top 3 candidates

    # Create a DataFrame from the top 3 candidates
    #df = pd.DataFrame(top_3_candidates, columns=["Candidate Name", "Party Name", "Total Votes"])
    
    list_name = [first_party_name,second_party_name,difference,constituency]

    
    #df = pd.DataFrame(list_name, columns=["First Party", "Second Party", "Difference","Constituency"])
    df["First Party"] = [first_party_name]
    df["Second Party"] = [second_party_name]
    df["Difference"] = [difference]
    df["Constituency"] = [constituency]
    
    
    # Print the DataFrame
    return list_name
```

```python
party(1)
```

```python

list_result = []
for i in range(1,224):
    list_result.append(party(i))
    
    
```

```python
df2 = pd.DataFrame(list_result,columns=["First Party", "Second Party", "Difference","Constituency"])
```
